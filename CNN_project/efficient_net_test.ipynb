```python
# split to train/val/test
import random
import shutil

# Define the path to your dataset folder
dataset_path = "spectrograms_noise"

# Define the paths for the train, validation, and test folders
train_path = "train"
val_path = "validate"
test_path = "test"

# Define the percentage splits
train_split = 0.7
val_split = 0.2
test_split = 0.1

# Create the train, validation, and test folders if they don't exist
os.makedirs(train_path, exist_ok=True)
os.makedirs(val_path, exist_ok=True)
os.makedirs(test_path, exist_ok=True)

# Loop through the subfolders (fake and real)
for class_name in ["fakes", "reals"]:
    # Get the list of images in the current subfolder
    subfolder_path = os.path.join(dataset_path, class_name)
    image_list = os.listdir(subfolder_path)
    random.shuffle(image_list)  # Shuffle the list of images

    # Calculate the number of images for each split
    num_images = len(image_list)
    num_train = int(train_split * num_images)
    num_val = int(val_split * num_images)
    num_test = int(test_split * num_images)

    # Split the images into train, validation, and test sets
    train_images = image_list[:num_train]
    val_images = image_list[num_train:num_train + num_val]
    test_images = image_list[num_train + num_val:]

    # Move the images to their respective folders
    for image_name in train_images:
        src = os.path.join(subfolder_path, image_name)
        dst = os.path.join(train_path, class_name, image_name)
        os.makedirs(os.path.join(train_path, class_name), exist_ok=True)
        shutil.copy(src, dst)

    for image_name in val_images:
        src = os.path.join(subfolder_path, image_name)
        dst = os.path.join(val_path, class_name, image_name)
        os.makedirs(os.path.join(val_path, class_name), exist_ok=True)
        shutil.copy(src, dst)

    for image_name in test_images:
        src = os.path.join(subfolder_path, image_name)
        dst = os.path.join(test_path, class_name, image_name)
        os.makedirs(os.path.join(test_path, class_name), exist_ok=True)
        shutil.copy(src, dst)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader

from torchvision.models import efficientnet_b7, EfficientNet_B7_Weights
```

```python
# Set the device (GPU if available, else CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
transform=transforms.Compose([
                                transforms.Resize(256),
                                transforms.CenterCrop(256),
                                transforms.ToTensor(),
                                transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
                           ])
```

```python
# Define paths to train, validation, and test datasets
train_path = "train"
val_path = "validate"
test_path = "test"

# Create data loaders
train_dataset = ImageFolder(train_path, transform=transform)
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)

val_dataset = ImageFolder(val_path, transform=transform)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)

test_dataset = ImageFolder(test_path, transform=transform)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
```

```python
# Build the EfficientNet-B7 model
model = efficientnet_b7(weights = EfficientNet_B7_Weights)
num_ftrs = model.classifier[1].in_features
model.classifier[1] = nn.Linear(num_ftrs, 2)  # Two output classes (fake or real)
model = model.to(device)
```

```python
# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python

# Training loop
num_epochs = 5
best_val_acc = 0.0

for epoch in range(num_epochs):
    train_loss = 0.0
    model.train()
    
    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(images)
        
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item() * images.size(0)
    
    train_loss = train_loss / len(train_dataset)

    # Validation
    model.eval()
    val_correct = 0
    val_total = 0
    
    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)
            
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            
            val_total += labels.size(0)
            val_correct += (predicted == labels).sum().item()
    
    val_acc = val_correct / val_total
    
    print(f"Epoch [{epoch+1}/{num_epochs}] - Train Loss: {train_loss:.4f}, Val Acc: {val_acc:.4f}")
    
    # Save the best model
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), "best_model.pth")
```
